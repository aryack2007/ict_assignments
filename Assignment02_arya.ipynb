1. Fizzbuzz

```python
for i in range(1,101):
  if i%3==0 and i%5==0:
    print(i,'= fizzbuzz')
  elif i%3==0:
   print(i,'= fizz')
  elif i%5==0:
    print(i,'= Buzz')
  else:
    print(i)
```

2. Electricity bill calculation

```python
unit = float(input('Enter number of units consumed: '))
bill_amount = 0
if unit<=100:
  bill_amount = unit*5
elif unit<=200:
  bill_amount = (100*5)+(unit-100)*7
else:
  bill_amount = (100*5)+(100*7)+((unit-200)*10)
  print('total bill amount = ₹',bill_amount)

```

4. Palindrome

```python
p = input('Enter num/str to be checked: ')
rev = p[::-1]
if p == rev:
   print("given num/str is a palindrome")
else:
   print("Given num/str is not palindrome")
```
